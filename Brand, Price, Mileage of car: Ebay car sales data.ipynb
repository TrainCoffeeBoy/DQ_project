I investigate the relationship between brand and price below:

```python
#Brand
autos['brand'].unique()
```

```python
autos['brand'].value_counts(normalize=True)
```

I choose top 6 brand (percentage of the total values > 5%) to aggregate

```python
brand_dict = {}

for b in range(0,6):
    price = autos[autos['brand'] == autos['brand'].value_counts().index[b]]['price'].mean()
    brand_dict[autos['brand'].value_counts().index[b]] = round(price)
    
brand_dict
```

opel is cheapest. audi, bmv and mercedes are expensive.

```python
mile_dict = {}

for b in range(0,6):
    mileage = autos[autos['brand'] == autos['brand'].value_counts().index[b]]['odometer_km'].mean()
    mile_dict[autos['brand'].value_counts().index[b]] = round(mileage)
    
mile_dict
```

```python
bmp_series = pd.Series(brand_dict)
bmm_series = pd.Series(mile_dict)
```

```python
df = pd.DataFrame(bmp_series, columns=['mean_price'])
df['mean_mileage'] = bmm_series
```
