```python
def update_vals(x):
    if pd.isnull(x):
        x = np.nan
    elif x == '-':
        x = False
    else:
        x = True       
    return x
```

```python
def inst_modify(x):
    if x < 1:
        x = 'Less than 1 year'
    elif x < 3:
        x = '1-2'
    elif x < 5:
        x ='3-4'
    elif x < 7:
        x = '5-6'
    elif x < 11:
        x = '7-10'
    elif x < 21:
        x = '11-20'
    else:
        x = 'More than 20 years'
        return x
```

```python
dete_resignations_up['institute_service_modified'] = dete_resignations_up['institute_service'].apply(inst_modify)
```

```python
combined_updated['institute_service'].astype('str').str.replace('Less than ','')\
.str.replace('More than ','').str.replace(' years','').str.replace('year','')\
.str.replace('11-','').str.replace(r'[0-9]-','').astype('float').value_counts()
```

```python
def carrer(x):
    if pd.isnull(x):
        return
    elif x < 3:
        x = 'New'
        return x
    elif x < 7:
        x = 'Experienced'
        return x
    elif x < 11:
        x = 'Established'
        return x
    else:
        x = 'Veteran'
        return x
```

```python
service_cat = combined_updated['institute_service'].astype('str').str.replace('Less than ','')\
.str.replace('More than ','').str.replace(' years','').str.replace('year','')\
.str.replace('11-','').str.replace(r'[0-9]-','').astype('float').apply(carrer)
```
